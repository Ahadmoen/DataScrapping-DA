# WEB SCRAPPING (AMAZON - GAMING SECTION)

```python
from bs4 import BeautifulSoup
import requests
```

```python
import requests

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Connection": "keep-alive"
}

response = requests.get(
    'https://www.amazon.com/s?k=gaming&_encoding=UTF8&content-id=amzn1.sym.5b60f7cf-8cdf-42b5-80ee-74f9ab340b5d&pd_rd_r=a6a20aa3-fa9f-4283-9f9a-5aaae2f55e0b&pd_rd_w=6huXa&pd_rd_wg=u55UP&pf_rd_p=5b60f7cf-8cdf-42b5-80ee-74f9ab340b5d&pf_rd_r=TG856Q7CXCJBNPTN11Z2&ref=pd_hp_d_atf_unk',
    headers=headers
)

print(response)
```

```python
soup = BeautifulSoup(response.text, 'lxml')
#print(soup.prettify()[:500])
```

```python
gaming_products = soup.find_all('div', class_ = 'sg-col-4-of-24 sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col s-widget-spacing-small sg-col-4-of-20 gsx-ies-anchor')
```

```python

for gaming_product in gaming_products:
    title = gaming_product.find('div', class_= "a-section a-spacing-none a-spacing-top-small s-title-instructions-style")
    #print(title.text)
    
    price = gaming_product.find('span', class_ = ['a-price-whole', 'a-price-decimal'])
    #print(price.text)
    
    rating_count = gaming_product.find('span', class_ = 'a-size-base s-underline-text')
    #print(rating_count.text)
       
    
    delivery_time = gaming_product.find('span', class_ = 'a-color-base a-text-bold') 
    #print(delivery_time.text)
    
    details =  gaming_product.find('div', class_=['a-fixed-left-grid', 'AverageCustomerReviews', 'a-spacing-small'])
    # print(details.prettify())
    
    rating = details.find('span', class_="a-icon-alt").text.split()[0]
    #print(rating)
    
    print(f'Title : {title.text}')
    #print(f'details : {details.text}')
    if price is not None:
        print(f'Price : {price.text}')
    else:
        print(f'Price : {0.0}')
    
    print(f'RatingCount : {rating_count.text}')
    if delivery_time is not None:
       print(f'DeliveryTIme : {delivery_time.text}')
    else:
        print(f'DeliveryTIme : {None}')
    print(f'rating : {rating}')
    
    print(' ')
    
    
```
